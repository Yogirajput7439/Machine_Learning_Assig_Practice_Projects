```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme() # set default theme - white, whitegrid, darkdark, dark, ticks
```

```python
tips = sns.load_dataset("tips")
```

```python
print(tips)
```

```python
sns.scatterplot(
    data = tips,
    x= "total_bill",
    y= "tip",
    hue= "time"
    )
```

```python
flights = sns.load_dataset("flights")
```

```python
flights.head()
```

```python
sns.lineplot(
                data=flights,
               x = "year",
               y = "passengers",
               )
```

```python
# categorical plot
# Bar Plots

tips.head()
```

```python
# in the seaborn data will automatically divide the male and female
  
sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex"
)
```

```python
sns.boxplot(
    data=tips,
    x="day",
    y="tip"
    
)
```
